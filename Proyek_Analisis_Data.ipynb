## Analisis Lanjutan (Opsional)

### Binning Berdasarkan Transaksi

```python
# membuat bins untuk transaction value
bins = [0, 100, 500, float('inf')]  # Define the bin ranges
labels = ['Small Transaction', 'Medium Transaction', 'Large Transaction']  # Labels for each bin

# menerapkan binning pada kolom payment_value
df_payments['transaction_category'] = pd.cut(df_payments['payment_value'], bins=bins, labels=labels, right=False)

# mengelompokkan transaksi
transaction_binning = df_payments.groupby('transaction_category').size()

# menampilkan visualisasi
colors = plt.cm.Greens(np.linspace(0.2, 0.8, len(transaction_binning))) 

# Visualisasikan hasil binning dengan gradasi warna hijau
plt.figure(figsize=(8, 6))
transaction_binning.plot(kind='bar', color=colors)

plt.title('Distribusi Transaksi Berdasarkan Kategori Nilai')
plt.xlabel('Kategori Transaksi')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()

transaction_binning
```

**Insight**
- Transaksi Kecil Mendominasi: Kategori Small Transaction (< 100) adalah yang paling banyak terjadi, dengan lebih dari 51.000 transaksi. Ini menunjukkan bahwa sebagian besar transaksi di marketplace adalah transaksi dengan nilai kecil.

- Transaksi Sedang Cukup Tinggi: Kategori Medium Transaction (100 - 500) juga memiliki jumlah transaksi yang cukup tinggi, yaitu sekitar 47.768 transaksi. Ini menandakan bahwa transaksi bernilai menengah juga cukup umum di platform ini.

- Transaksi Besar Relatif Sedikit: Kategori Large Transaction (> 500) memiliki jumlah yang jauh lebih sedikit, hanya 4.263 transaksi. Ini bisa menunjukkan bahwa pembelian dengan nilai besar relatif jarang terjadi di platform ini.
